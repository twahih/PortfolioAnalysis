```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import edhec_risk_kit as erk
import pandas as pd
import numpy as np
```

```python
ind_return=erk.get_ind_returns()
tmi_return=erk.get_total_market_index_returns()
```

```python
risky_r = ind_return["2000":][["Steel","Fin","Beer"]]
#Safe Asset
safe_r=pd.DataFrame().reindex_like(risky_r)
```

```python
safe_r[:]=0.03/12
start=1000
floor=0.8
```

1. Cushion -(Asset Value minus Floor Value)
2. Compute an allocation to safe and risky assets  -->m*risk budget
3. Recompute the Asset value based on the returns 


```python
dates=risky_r.index
n_steps = len(dates)
account_value=start
floor_value=start*floor
m=3
account_history=pd.DataFrame().reindex_like(risky_r)
cushion_history=pd.DataFrame().reindex_like(risky_r)
risky_w_history=pd.DataFrame().reindex_like(risky_r)
```

```python
for step in range(n_steps):
    cushion=(account_value - floor_value)/account_value
    risky_w=m*cushion
    risky_w=np.minimum(risky_w,1)
    risky_w=np.maximum(risky_w,0)
    safe_w=1-risky_w
    risk_alloc=account_value*risky_w
    safe_alloc=account_value*safe_w
    # Update the account vlaue for this step 
    account_value=risk_alloc*(1+risky_r.iloc[step])+safe_alloc*(1+safe_r.iloc[step])
    cushion_history.iloc[step]=cushion
    risky_w_history.iloc[step]=risky_w
    account_history.iloc[step]=account_value
```

```python
risky_wealth=start*(1+risky_r).cumprod()
```

```python
ax=account_history["Fin"].plot(figsize=(12,6))
risky_wealth["Fin"].plot(ax=ax,style="k:",label="risky",legend=True)
ax.axhline(y=floor_value,color='r',linestyle='--')
```
